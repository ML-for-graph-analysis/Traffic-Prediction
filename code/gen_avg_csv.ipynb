Averaged speed for 3 time slots below per day
* 6 - 10
* 10- 16
* 16 - 21

```python
%pwd
```

```python
import pandas as pd

df = pd.read_csv('../data/la_speed.csv')
```

```python
ids = df.columns
```

```python
out_df = pd.DataFrame(columns=ids) 
out_df
```

```python
days_list = df.iloc[:, 0].str.split(" ").str[0].unique().tolist()
# days_list
```

```python
for day in days_list:
    day_df = df.loc[df.iloc[:, 0].str.split(" ").str[0] == day]
    
    morning = day_df.loc[df.iloc[:, 0].str.split(" ").str[1].str.split(":").str[0].isin(['06', '07', '08', '09'])].mean(axis=0).round(2)
    regular = day_df.loc[df.iloc[:, 0].str.split(" ").str[1].str.split(":").str[0].isin(['10', '11', '12', '13', '14', '15'])].mean(axis=0).round(2)
    evening = day_df.loc[df.iloc[:, 0].str.split(" ").str[1].str.split(":").str[0].isin(['16', '17', '18', '19', '20'])].mean(axis=0).round(2)
    
    avg_morning = morning.append(pd.Series([day + ' ' + '06:00:00 - 09:55:00']))
    avg_regular= regular.append(pd.Series([day + ' ' + '10:00:00 - 15:55:00']))
    avg_evening = evening.append(pd.Series([day + ' ' + '16:00:00 - 20:55:00']))
    
    out_df = out_df.append(avg_morning, ignore_index=True)
    out_df = out_df.append(avg_regular, ignore_index=True)
    out_df = out_df.append(avg_evening, ignore_index=True)
    
```

```python
cols = out_df.columns.tolist()
cols = cols[-1:] + cols[:-1]
out_df = out_df[cols] 
```

```python
out_df.drop(df.filter(regex="Unname"),axis=1, inplace=True)
out_df.rename(columns = {0: "time"}, inplace=True)
out_df
```
